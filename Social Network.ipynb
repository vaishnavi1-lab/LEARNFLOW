# Network Construction

```python
import networkx as nx
```

```python
G = nx.DiGraph()

# Add nodes
G.add_nodes_from(sn['User ID'].unique())
```

```python
# Add edges
edges = list(zip(sn['User ID'], sn['Mentions'].str.split(',')))
for edge in edges:
    source, targets = edge[0], edge[1]
    for target in targets:
        G.add_edge(source, target.strip())
```

```python
# Visualize the network (optional)
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=50, font_size=8, font_color='black', alpha=0.7, node_color='skyblue', edge_color='gray')
plt.title("Social Network Graph")
plt.show()
```

# Identifying Key Metrices

```python
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=500, tol=1e-6)

# Add centrality measures to the DataFrame
sn['Degree Centrality'] = sn['User ID'].map(degree_centrality)
sn['Betweenness Centrality'] = sn['User ID'].map(betweenness_centrality)
sn['Eigenvector Centrality'] = sn['User ID'].map(eigenvector_centrality)

# Display the DataFrame with centrality measures
print("DataFrame with Centrality Measures:")
print(sn[['User ID', 'Degree Centrality', 'Betweenness Centrality', 'Eigenvector Centrality']])

# Identify influential users based on centrality measures (customize threshold as needed)
influential_users = sn[sn['Degree Centrality'] > 0.05]['User ID'].tolist()

# Display influential users
print("\nInfluential Users:")
print(influential_users)
```

# Visualization

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from(sn['User ID'].unique())

# Add edges
edges = list(zip(sn['User ID'], sn['Mentions'].str.split(',')))
for edge in edges:
    source, targets = edge[0], edge[1]
    for target in targets:
        G.add_edge(source, target.strip())

# Calculate degree centrality for node sizes
degree_centrality = nx.degree_centrality(G)
node_sizes = [2000 * degree_centrality[node] for node in G.nodes()]

# Identify influential users based on degree centrality threshold
influential_users = [user_id for user_id in G.nodes() if degree_centrality[user_id] > 0.05]

# Create a visualization
plt.figure(figsize=(12, 8))

# Draw the network graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=node_sizes, font_size=8, font_color='black', alpha=0.7, node_color='skyblue', edge_color='gray')

# Highlight influential users with a different color
nx.draw_networkx_nodes(G, pos, nodelist=influential_users, node_color='salmon', node_size=3000)

# Add labels and title
plt.title("Social Network Graph with Influential Users")
plt.show()
```

# Interpretation and Insights

```python
print("Interpretation and Insights:")
print("\n1. **Influential Users:**")
print("   - Users with a degree centrality greater than 0.05 are considered influential.")
print(f"   - Identified influential users: {influential_users}")

# Community Structures
print("\n2. **Community Structures:**")
communities = list(nx.weakly_connected_components(G))
for i, community in enumerate(communities, 1):
    print(f"   - Community {i}: {list(community)}")

# Popular Hashtags and Mentions
print("\n3. **Popular Hashtags and Mentions:**")
hashtags = sn['Hashtags'].str.split(',').explode().str.strip().value_counts().head(5)
mentions = sn['Mentions'].str.split(',').explode().str.strip().value_counts().head(5)

print("   - Top 5 Hashtags:")
print(hashtags)

print("\n   - Top 5 Mentions:")
print(mentions)

# Temporal Analysis
print("\n4. **Temporal Analysis:**")
sn['Post Timestamp'] = pd.to_datetime(sn['Post Timestamp'])
sn['Month'] = sn['Post Timestamp'].dt.month

monthly_post_counts = sn.groupby('Month').size()

plt.figure(figsize=(12, 6))
monthly_post_counts.plot(kind='bar', color='skyblue')
plt.title("Monthly Post Distribution")
plt.xlabel("Month")
plt.ylabel("Number of Posts")
plt.show()
```
